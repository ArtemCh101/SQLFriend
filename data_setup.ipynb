```python
import kagglehub
import os
import shutil

download_path = kagglehub.dataset_download("jeromeblanchet/yale-universitys-spider-10-nlp-dataset")
print(f"Source path: {download_path}")

dest_path = os.getcwd()

def consolidate_spider_data(src_root, dst_node):
    target_files = ['tables.json', 'train_spider.json', 'dev.json', 'train_others.json']
    target_folders = ['database']
    
    found_items = 0
    
    for root, dirs, files in os.walk(src_root):
        for filename in files:
            if filename in target_files:
                src_file = os.path.join(root, filename)
                shutil.copy2(src_file, os.path.join(dst_node, filename))
                print(f"Copied file: {filename}")
                found_items += 1
        
        for dirname in dirs:
            if dirname in target_folders:
                src_dir = os.path.join(root, dirname)
                dest_dir = os.path.join(dst_node, dirname)
                if os.path.exists(dest_dir):
                    shutil.rmtree(dest_dir)
                shutil.copytree(src_dir, dest_dir)
                print(f"Copied directory: {dirname}/")
                found_items += 1

    return found_items

items_moved = consolidate_spider_data(download_path, dest_path)

if items_moved > 0:
    print(f"\nSuccess: {items_moved} assets moved to {dest_path}")
    print("You can now access 'tables.json', 'train_spider.json', and the 'database/' folder directly.")
else:
    print("\nError: Could not locate Spider assets in the downloaded path.")

print("\nLocal directory content:")
print(os.listdir(dest_path))
```
